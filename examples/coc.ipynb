```python
from scipy.optimize import least_squares
```

```python
M = idxs.values
M
```

```python
target = cWar.getClanAttacks().destructionPercentage / 100
target
```

```python
x0 = np.zeros(shape=(idxs.shape[1]))
x0
```

```python
def funzione(x, M, target):
    """ La funzione da minimizzare:
        M*x - y = 0
    """
    return np.dot(M, x) - target
```

```python
res = least_squares(funzione, x0, args=[M, target], bounds=[-1,1])
res
```

```python
pd.concat([pd.DataFrame(np.dot(M, res.x)), pd.DataFrame(target)], axis=1)
```
